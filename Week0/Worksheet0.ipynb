```python
# Daily time (in hours): [study, entertainment, sleep]
time_data = [
(3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
(4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
(5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
(3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
(4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]
```

```python
#Task7.1
low = []
moderate = []
high = []

for i in time_data: 
    if i[0] < 3:
        low.append(i)
    elif i[0] >= 3 and i[0] <= 5:
        moderate.append(i)
    else:
        high.append(i)

print("The time data with low study time: ", low)
print("The time data with moderate study time: ", moderate)
print("The time data with high study time: ", high)
```

```python
#Task7.2

print("The number of days with low study time is: ", len(low))
print("The number of days with moderate study time is: ", len(moderate))
print("The number of days with high study time is: ", len(high))
```

```python
#Task7.3

study_minutes = []

for i in time_data:
    study_minutes.append(i[0] * 60)

print("The converted hours into minutes are: ", study_minutes)
```

```python
#Task7.4

study_hours = []
entertainment_hours = []
sleep_hours = []

for i in time_data:
    study_hours.append(i[0])
    entertainment_hours.append(i[1])
    sleep_hours.append(i[2])

print(f"The average hour spent studying: {sum(study_hours)/len(study_hours)} hours")
print(f"The average hour spent on entertainment: {sum(entertainment_hours)/len(entertainment_hours)} hours")
print(f"The average hour spent sleeping: {sum(sleep_hours)/len(sleep_hours):.2f} hours")
```

```python
#Task7.5

import matplotlib.pyplot as plt

plt.scatter(study_hours, sleep_hours, color="red")
plt.xlabel("Study Hours")
plt.ylabel("Sleep Hours")
plt.title("Study vs Sleep Pattern:")
```

```python
#8.1.1 Task1.1
def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)
        else:
            total += element

    return total

#Task1.2
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
print(sum_nested_list(nested_list))
```

```python
#8.1.1 Task2
def generate_permutations(s):
    if len(s) == 0:
        return [""]
    permutations = set()
    for i in range(len(s)):
        current = s[i]
        remaining = s[:i]+s[i+1:]
        for p in generate_permutations(remaining):
            permutations.add(current + p)
    return permutations

print(list(generate_permutations("abc")))
print(list(generate_permutations("aab")))
```

```python
#8.1.1 Task3.1
def calculate_directory_size(directory):
    total = 0
    for item in directory.values():
        if isinstance(item, dict):
            total += calculate_directory_size(item)
        else:
            total += item
    return total
```

```python
#8.2.1 Task1.1
def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    return dp[amount] if dp[amount] != float('inf') else -1

#Task1.2
print(min_coins([1,2,5],11))
```

```python
#8.2.1 Task2.1
def longest_common_subsequence(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[m][n]
    
#Task2.2
print(longest_common_subsequence("abcde", "ace"))
```

```python
#8.2.1 Task3.1
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(values[i - 1] + dp[i - 1][w - weights[i - 1]], dp[i - 1][w])
            else:
                dp[i][w] = dp[i - 1][w]
    
    return dp[n][capacity]

#Task3.2
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
print(knapsack(weights, values, capacity))
```
